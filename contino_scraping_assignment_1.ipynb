```python
base_url = "https://www.federalregister.gov/api/v1/documents.json?per_page=20&conditions[docket_id]="
```

```python
broken_endpoints = [] 
incidents_list = []
for code in disaster_codes:
    endpoint = base_url + code
    try:
        response = requests.get(endpoint)
        data = response.json()
        incident_text = data.get("results")[0].get("abstract")
        state_text = data.get("results")[0].get("title")
        incidents_list.append({"disaster_code": code,
                              "description": incident_text,
                              "state": state_text})
    except:
        broken_endpoints.append(endpoint)
```

```python
actual_incidents = [item.get("description") for item in incidents_list]
actual_incidents
```

```python
import regex as re
incident_names = []

for incident in actual_incidents:
    pat1 = re.compile(r"incident: (\D*)\.\sIncident Period:", re.I)
    pat_results = pat1.findall(incident)
    #print(pat_results)

    incident_names.append(pat_results)
incident_names
```

```python
incidents = [item for sublist in incident_names for item in sublist]

print(incidents)
```

```python
actual_states = [item.get("state") for item in incidents_list]
actual_states
```

```python
type(actual_states)
```

```python
state_names = []

for state in actual_states:
    pat2 = re.compile(r"the (\D* of \D*)", re.I)
    pat2_results = pat2.findall(state)
    print(pat2_results)
    state_names.append(pat2_results) 
state_names
```

```python
states = [
   item.replace("State of ", "").replace("Commonwealth of ", "") if isinstance(item, str) else "Nested item"
   for sublist in state_names
    for item in sublist
]

print(states)
```

```python
max_length = max(len(disaster_codes), len(states), len(incidents))
disaster_codes += [None] * (max_length - len(disaster_codes))
states += [None] * (max_length - len(states))
incidents += [None] * (max_length - len(incidents))

# Create the DataFrame
final_incidents = {
    "Disaster code": disaster_codes,
    "State": states,
    "Incident": incidents
}

df = pd.DataFrame(final_incidents)

print(df)
```
